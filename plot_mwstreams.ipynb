```python
import numpy as np
import scipy
import pylab as plt
import mypyutils as mylib
import myutils
import bovy_coords as bovyc
import mwstreams as mwst
import gcutils 
%pylab inline
```

```python
fig=plt.figure(1,figsize=(14,8))
ax1=fig.add_subplot(111)

my_cmap=myutils.get_sron_rainbow(N=12)
Rmax=30.
galactic=True #set to False for equatorial

#All MW known streams
mwst.plot_grillmair_carlin_streams(ax1,Rstat='median',scat_kwargs=dict(vmin=0.,vmax=Rmax,cmap=my_cmap,alpha=0.3),text_kwargs=dict(fontsize=13),
                                   galactic=galactic,plot_stream_type='all',plot_names=True,Nran=1000)
       
#Globular clusters-----------------
mwst.plot_globular_clusters(ax1,scat_kwargs=dict(vmin=0.,vmax=Rmax,cmap=my_cmap),galactic=galactic)        

#PS-streams------------------------
cc=mwst.plot_PS1_streams(ax1,scat_kwargs=dict(s=10,vmin=0.,vmax=Rmax,cmap=my_cmap,zorder=15),
                         text_kwargs=dict(fontsize=13,zorder=15),sym_kwargs=dict(zorder=15),galactic=galactic)
cbar = plt.colorbar(cc,ax=ax1,pad=0.01,fraction=0.05,aspect=30)    

#Orphan stream
mwst.plot_orphan_stream(ax1,scat_kwargs=dict(marker='H',s=80,vmin=0.,vmax=Rmax,edgecolor='none',cmap=my_cmap,zorder=15),
                        text_kwargs=dict(fontsize=13,zorder=16),galactic=galactic)     
    

#=======================plot style===========================================================
yticks=plt.MultipleLocator(20.)
ax1.yaxis.set_major_locator(yticks)
xticks=plt.MultipleLocator(30.)
ax1.xaxis.set_major_locator(xticks)

ax1.grid()
ax1.set_xlim(-5.,360.)
ax1.set_ylim(-90,90.)
#ax1.set_xlim(120,250.)
#ax1.set_ylim(-50,10.)
ax1.set_xlabel('l ($^\circ$)')
ax1.set_ylabel('b ($^\circ$)')

#plt.savefig('known_streams.lb.fermi_groups_rrls_%s.png' % (lab))
```
