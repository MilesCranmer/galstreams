```python
import numpy as np
import scipy
import pylab as plt
import mwstreams
import myutils
%pylab inline
```

```python
mwsts=mwstreams.MWStreams(verbose=False,gcstep=0.05,N=8000)
nnot=['Hermus','Hyllus','Lethe','Cocytos','Acheron','Alpheus','Scamander','Sangarius']
for i in mwsts.keys():
    if i in nnot : mwsts[i].name=mwsts[i].name[:3]
    if mwsts[i].phi is not None:    
     mwsts[i].phi= (mwsts[i].phi+180.) % 360.
     mwsts[i].cphi= (mwsts[i].cphi+180.) % 360.   
```

```python
print mwsts['Sgr-L00'].ra.size
mask=np.arange(mwsts['Sgr-L00'].ra.size)[::20]
mwsts['Sgr-L00'].mask_footprint(mask)
mwsts['Sgr-L00'].cphi,mwsts['Sgr-L00'].ctheta=350.,-15.
```

```python
plt.rc('font', size=14)
plt.rc('axes', labelsize=18)
plt.rc('xtick', labelsize=16)
plt.rc('ytick', labelsize=16)

fig=plt.figure(1,figsize=(16,8))
ax=fig.add_subplot(111)
my_cmap=myutils.get_sron_rainbow(N=12)
my_cmap.set_under('gray')
ckws=dict(cmap=my_cmap,vmax=50.,vmin=0.)
colorbar_kwargs=dict(pad=0.01,label='Mean $R_{\mathrm{gal}}$ (kpc)',aspect=30.)
mwsts.plot_stream_compilation(ax,scat_kwargs=ckws,verbose=True,plot_colorbar=True,cootype='GC',
                              cb_kwargs=colorbar_kwargs,exclude_streams=['Her-Aq'])#,Rrange=[0.,30.])
ax.grid()
ax.set_xlabel('$\phi$ ($^\circ$)')
ax.set_ylabel('$\\theta$ ($^\circ$)')
ax.set_xlim(0.,360.)
ax.set_ylim(-90,90.)
majorLocator=plt.MultipleLocator(15.)
ax.yaxis.set_major_locator(majorLocator)
plt.savefig('MWStreams_lib.galactocentric.png')
```
